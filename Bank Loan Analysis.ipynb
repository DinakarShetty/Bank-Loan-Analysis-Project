## Total Loan Applications

```python
Total_loan_Application= df['id'].count()
Total_loan_Application
```

## MTD Total Loan Applications

```python
Latest_issue_date=df['issue_date'].max()
Latest_year=Latest_issue_date.year
Latest_month=Latest_issue_date.month

mtd_data=df[(df['issue_date'].dt.year==Latest_year) & (df['issue_date'].dt.month==Latest_month)]
mtd_loan_application=mtd_data['id'].count()
print(f"mtd_loan_pplication(for {Latest_issue_date.strftime('%B %Y')}):{mtd_loan_application}")
```

## Total Funded Amount

```python
Total_funded_amount=df['loan_amount'].sum()
Total_funded_amount_millions=Total_funded_amount/1000000
print("Total Funded Amount: ${:.2f}M" .format(Total_funded_amount_millions))
```

## MTD Total Funded Amount

```python
Latest_issue_date=df['issue_date'].max()
Latest_year=Latest_issue_date.year
Latest_month=Latest_issue_date.month

mtd_data=df[(df['issue_date'].dt.year==Latest_year) & (df['issue_date'].dt.month==Latest_month)]
MTD_Total_funded_amount=mtd_data['loan_amount'].sum()
MTD_Total_funded_amount_millions=MTD_Total_funded_amount/1000000
print("MTD_Total_funded_amount : ${:.2f}M" .format(MTD_Total_funded_amount_millions))
```

## Total Amount Received

```python
Total_amount_Received=df['total_payment'].sum()
Total_amount_Received_millions=Total_amount_Received/1000000
print("Total Funded Amount: ${:.2f}M" .format(Total_amount_Received_millions))
```

## MTD_Total Amount Received

```python
Latest_issue_date=df['issue_date'].max()
Latest_year=Latest_issue_date.year
Latest_month=Latest_issue_date.month

mtd_data=df[(df['issue_date'].dt.year==Latest_year) & (df['issue_date'].dt.month==Latest_month)]
MTD_Total_amount_Received=mtd_data['total_payment'].sum()
MTD_Total_amount_Received_millions=MTD_Total_amount_Received/1000000
print("MTD_Total_amount_Received : ${:.2f}M" .format(MTD_Total_amount_Received_millions))
```

## Average Interest Rate

```python
Average_interest_rate=df['int_rate'].mean()*100
print("Average interest Rate is: ${:.2f}% ".format(Average_interest_rate))
```

## Average Debt-To-Income Ratio(DTI)

```python
Average_DTI=df['dti'].mean()
print("Average DTI is: ${:.2f}% ".format(Average_DTI))
```

## Good Loan Matrix

```python
Good_loan=df[df['loan_status'].isin(['Fully Paid','Current'])]

Total_loan_Application= df['id'].count()

Good_loan_applications=Good_loan['id'].count()
Good_loan_funded_amount=Good_loan['loan_amount'].sum()
Good_loan_funded_amount_millions=Good_loan_funded_amount/1000000
Good_loan_received=Good_loan['total_payment'].sum()
Good_loan_received_millions=Good_loan_received/1000000
Good_loan_percentage=(Good_loan_applications/Total_loan_Application)*100

print("Good loan applications is " ,Good_loan_applications)
print("Good loan funded amount is (millions) ${:.2f}M ".format(Good_loan_funded_amount_millions))
print("Good loan received is (millions) ${:.2f}M ".format(Good_loan_received_millions))
print("Good loan percentage is {:.2f}% ".format(Good_loan_percentage))
```

## Bad Loan Matrix

```python
Bad_loan=df[df['loan_status'].isin(['Charged Off'])]

Total_loan_Application= df['id'].count()

Bad_loan_applications=Bad_loan['id'].count()
Bad_loan_funded_amount=Bad_loan['loan_amount'].sum()
Bad_loan_funded_amount_millions=Bad_loan_funded_amount/1000000
Bad_loan_received=Bad_loan['total_payment'].sum()
Bad_loan_received_millions=Bad_loan_received/1000000
Bad_loan_percentage=(Bad_loan_applications/Total_loan_Application)*100

print("Bad loan applications is " ,Bad_loan_applications)
print("Bad loan funded amount is (millions) ${:.2f}M ".format(Bad_loan_funded_amount_millions))
print("Bad loan received is (millions) ${:.2f}M ".format(Bad_loan_received_millions))
print("Bad loan percentage is {:.2f}% ".format(Bad_loan_percentage))
```
